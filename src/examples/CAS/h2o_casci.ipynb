```python
import numpy as np
import sys
#np.set_printoptions(precision=5, linewidth=200, suppress=True)
import psi4
from helper_PFCI import PFHamiltonianGenerator
from helper_PFCI import Determinant
from helper_cqed_rhf import cqed_rhf
np.set_printoptions(threshold=sys.maxsize)
```

```python
mol_tmpl = """
0 1
O
H 1 **R**
H 1 **R** 2 104.5
symmetry c1
"""

mol_str = """
0 1
O
H 1 1.6
H 1 1.6 2 104.5  
symmetry c1
"""
r_array = np.linspace(0.6, 2.0, 15)
CASCI_E_array = []
FCI_E_array = []
print(r_array)

options_dict = {'basis': 'sto-3g',
                  'scf_type': 'pk',
                  'e_convergence': 1e-10,
                  'd_convergence': 1e-10
                  }

#mol = psi4.geometry(mol_str)

cavity_options = {
    'omega_value' : 0.0,
    'lambda_vector' : np.array([0, 0, 0.0]),
    'ci_level' : 'cas',
    'ignore_coupling' : False,
    'number_of_photons' : 1,
    'natural_orbitals' : True,
    'davidson_roots' : 6,
    'davidson_maxdim':10,
    'rdm_weights' : np.array([1,1,1,1,1,1,1,1]),
    'nact_orbs' : 5,
    'nact_els' : 6,
    'full_diagonalization' : True,
    
}

psi4.set_options(options_dict)
#psi4.core.set_output_file('h2o_0.6.out', False)

for r in r_array:
    mol_str = mol_tmpl.replace("**R**", str(r))
    print(mol_str)
    mol = psi4.geometry(mol_str)
    #psi4.set_options({'restricted_docc': [3],'active': [4],'num_roots':2})
    scf_e, wfn = psi4.energy('SCF', return_wfn=True)
    fci_energy = psi4.energy('fci',ref_wfn=wfn)
    H2_PF = PFHamiltonianGenerator(mol_str, options_dict, cavity_options)
    CASCI_E_array.append(H2_PF.CIeigs[0])
    FCI_E_array.append(fci_energy)
    print(r, H2_PF.CIeigs[0])


## First compute SCF energy using Psi4
#scf_e, wfn = psi4.energy('SCF', return_wfn=True)

#psi4.set_options({'restricted_docc': [3],'active': [4],'num_roots':2})
#fci_energy = psi4.energy('fci',ref_wfn=wfn)
#print(np.allclose(H2_PF.cis_e[0],fci_energy,1e-8,1e-8))

```

```python
from matplotlib import pyplot as plt
plt.plot(r_array, CASCI_E_array, label='cisno 6,5')
plt.plot(r_array, FCI_E_array, label='fci')
#plt.plot(r_array, CCASCI_E_array, label='mo 6,5')
plt.legend()
plt.show()
```
